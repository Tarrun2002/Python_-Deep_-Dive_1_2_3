```python
def produce_element(dq,n):
  for i in range (1,n):
    dq.appendleft(i)
    if len(dq)==dq.maxlen:
      print("Queue full")
      yield 


def consume_element(dq):
  while True:
    while len(dq)>0:
      print("processing",dq.pop())
    print("queue empty")
    yield  

def coordinator():
  dq=deque(maxlen=10)
  producer=produce_element(dq,36)
  consumer =consume_element(dq)

  while True:
    try:
      print("Producing")
      next(producer)
    except StopIteration:
      break
    finally:
      print("consuming")
      next(consumer)
      




```

**Pipeline** :**Pushing** **Data**

```python
def coroutine (coro):
  def inner(*args,**kwargs):
    gen=coro(*args,**kwargs)
    next(gen)
    return gen
  return inner
```

```python
@coroutine
def handle_data():
  while True:
    received=yield
    print(received)
```

```python
import math

@coroutine
def power_up(n,next_gen):
  while True:
    recieved=yield
    output=math.pow(recieved,n)
    next_gen.send(output)
```

```python
print_data=handle_data()
gen=power_up(2,print_data)
```

```python
for i in range(1,6):
  gen.send(i)
```

```python
gen2=power_up(3,print_data)
gen1=power_up(2,gen2)
for i in range(1,6):
  gen1.send(i)
```

```python
@coroutine
def filter_even(next_gen):
  while True:
    recieved=yield
    if recieved %2==0:
      next_gen.send(recieved)
```

```python
filtered = filter_even(print_data)
gen2=power_up(3,filtered)
gen1=power_up(2,gen2)
for i in range(1,6):
  gen1.send(i)
```
